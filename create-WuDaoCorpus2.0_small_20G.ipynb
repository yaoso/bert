```python
def extract_subset(json_file):
    output = []
    tmp = json.load(open(json_file))
    for item in tmp:
        if item['dataType'] in dataType2prob:
            if np.random.random() < dataType2prob[item['dataType']]:
                output.append(item)
    if output:
        with open("/mnt/disks/persist/WuDaoCorpus2.0_small_20G/" + json_file.split('/')[-1], 'w',
                 encoding='utf-8') as fo:
            for item in output:
                fo.write(json.dumps(item).strip() + "\n")
#     return output
```

```python
outputs = Parallel(80)(delayed(extract_subset)(x) for x in tqdm(files))
```

```python
import spacy

nlp = spacy.load("zh_core_web_md")
```

```python
successful_text_filenames = set()  # set(['part-2021021808', 'part-2021022921'])
for text_file in text_files:
    successful_text_filenames.add(text_file.split('/')[-1][:-4])
```

```python
list(successful_text_filenames)[:3]
```

```python
candidate_small_files = []
for small_file in small_files:
    cur_filename = small_file.split('/')[-1][:-5]
    if cur_filename not in successful_text_filenames:
        candidate_small_files.append(small_file)
    
```

```python
len(candidate_small_files), len(successful_text_filenames), len(small_files)
```

```python
def sentence_segment(file_name):
    with open(file_name, 'r', encoding='utf-8') as fi, \
    open("/home/lj/WuDaoCorpus2.0_small_20G_text/" + file_name.split('/')[-1][:-5] + ".txt", 'w',
        encoding='utf-8') as fo:
        for line in fi:
            doc = json.loads(line.strip())
            nlp_output = nlp(doc['content'])
            for sent in nlp_output.sents:
                fo.write(sent.text.strip() + "\n")
            fo.write("\n")
```

```python
Parallel(50)(delayed(sentence_segment)(x) for x in tqdm(candidate_small_files))
```
